Data menunjukkan angka yang sangat ekstrem dan tidak realistis untuk produksi sampah harian. Sebagai contoh, Kabupaten Bogor tercatat menghasilkan 861919.33 ton sampah per hari dan Kabupaten Bandung mencapai 1156523.00 ton per hari. Angka-angka ini jelas tidak masuk akal untuk tingkat produksi sampah dalam skala kabupaten per harinya. Karena tidak masuk akal, jadinya dilakukan pembersihan data

```python
# Fungsi membersihkan data
def data_bersih(df, threshold_year=2020):
    df_clean = df.copy()

    # Identifikasi dan koreksi nilai yang sangat besar
    mask = (df_clean['tahun'] < threshold_year) & (df_clean['jumlah_sampah'] > 10000)
    df_clean.loc[mask, 'jumlah_sampah'] = df_clean.loc[mask, 'jumlah_sampah'] / 1000

    # Hitung jumlah data yang dikoreksi
    num_corrected = mask.sum()

    if num_corrected > 0:
      return df_clean

# Fungsi membandingkan plot
def plot_data_comparison(df_before, df_after):

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # Plot data sebelum pembersihan
    df_before.boxplot(column='jumlah_sampah', by='tahun', ax=ax1)
    ax1.set_title('Sebelum Pembersihan')

    # Plot data setelah pembersihan
    df_after.boxplot(column='jumlah_sampah', by='tahun', ax=ax2)
    ax2.set_title('Setelah Pembersihan')

    plt.tight_layout()
    plt.show()

# bersihkan data
jml_clean = data_bersih(jml)
```

Untuk mengatasi permasalahan tersebut, Pembersihan difokuskan pada data sebelum tahun 2020 yang memiliki nilai di atas 10.000 ton per hari. Metode yang digunakan adalah dengan membagi nilai-nilai tersebut dengan faktor 1000, yang menghasilkan angka yang lebih realistis dan konsisten dengan ekspektasi produksi sampah harian tingkat kabupaten/kota.

```python
# Menampilkan head dari data awal
print("=== Data Sebelum Pembersihan ===")
display(jml.head())

# Menampilkan head dari data setelah pembersihan
print("\n=== Data Setelah Pembersihan ===")
display(jml_clean.head())

# Panggil fungsi visualisasi
plot_data_comparison(jml, jml_clean)
```

Setelah proses pembersihan, terlihat perubahan yang signifikan pada dataset:

- Skala data mengalami penurunan dari jutaan ton menjadi ribuan ton per hari
- Kabupaten Bogor yang semula tercatat 861919.33 ton menjadi 861,91933 ton per hari
- Kabupaten Bandung yang awalnya 1156523.00 ton menjadi 1.156,523 ton per hari

```python
# Melakukan penggabungan data pada kolom 'id', 'kode_kabupaten_kota', 'nama_kabupaten_kota', dan 'tahun'
df_merged = pd.merge(pds, jml_clean, on=['id', 'kode_kabupaten_kota', 'nama_kabupaten_kota', 'tahun'], how='outer')

# Memberi nama kolom yang lebih deskriptif setelah gabungan
df_merged.rename(columns={
    'jumlah_produksi_sampah': 'jumlah_produksi_sampah',
    'jumlah_sampah': 'jumlah_sampah_ditangani'
}, inplace=True)

# Menambahkan kolom rasio_ditangani_terhadap_produksi
df_merged['rasio_ditangani_terhadap_produksi'] = df_merged['jumlah_sampah_ditangani'] / df_merged['jumlah_produksi_sampah'] * 100

# Memfilter data untuk tahun 2021 saja
df_2021 = df_merged[df_merged['tahun'] == 2021]

# Mengatur ulang urutan kolom
df_2021 = df_2021[['id', 'kode_kabupaten_kota', 'nama_kabupaten_kota', 'tahun', 'jumlah_produksi_sampah', 'jumlah_sampah_ditangani', 'rasio_ditangani_terhadap_produksi']]

# Menampilkan informasi data gabungan untuk tahun 2021
df_2021.info()
```

Nilai positif pada kolom ini menunjukkan adanya sampah yang tidak ditangani, sedangkan nilai negatif (yang tidak mungkin terjadi dalam konteks ini) akan menunjukkan kondisi sebaliknya.

6. **Mengatur Ulang Urutan Kolom**: Urutan kolom diatur ulang di DataFrame `df_2021` agar lebih terstruktur dan mudah dibaca.

5. Bagaimana hasil analisis korelasi ini dapat digunakan untuk meningkatkan alokasi sumber daya dan efektivitas pengangkutan sampah di masa mendatang?


```python
# Mengganti nama kolom untuk menghindari konflik
arm = arm.rename(columns={'nama_kabupaten_kota': 'nama_kabupaten_kota_arm', 'id': 'id_arm'})
truk = truk.rename(columns={'nama_kabupaten_kota': 'nama_kabupaten_kota_truk', 'id': 'id_truk'})
pck = pck.rename(columns={'nama_kabupaten_kota': 'nama_kabupaten_kota_pck', 'id': 'id_pck'})

# Menggabungkan dataset
df_sampah = pd.merge(jml_clean, mtr, on=['kode_kabupaten_kota', 'tahun'])
df_sampah = pd.merge(df_sampah, dump, on=['kode_kabupaten_kota', 'tahun'])
df_sampah = pd.merge(df_sampah, arm, on=['kode_kabupaten_kota', 'tahun'])
df_sampah = pd.merge(df_sampah, truk, on=['kode_kabupaten_kota', 'tahun'])
df_sampah = pd.merge(df_sampah, pck, on=['kode_kabupaten_kota', 'tahun'])

# Filter untuk tahun 2021
df_sampah_2021 = df_sampah[df_sampah['tahun'] == 2021]

# Menghitung korelasi
correlation = df_sampah_2021[['jumlah_sampah', 'jumlah_motor_sampah', 'jumlah_dumptruck', 'jumlah_armroll', 'jumlah_truck_sampah', 'jumlah_mobil_pickup']].corr()

# Visualisasi
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Korelasi antara Jumlah Sampah dan Alat Pengangkut (Tahun 2021)')
plt.show()
```

Heatmap ini menunjukkan korelasi antara jumlah sampah dan alat pengangkut sampah pada tahun 2021. Setiap sel dalam heatmap merepresentasikan nilai koefisien korelasi antara dua variabel. Berikut adalah penjelasan rinci dari korelasi yang terlihat:

1. `jumlah_sampah` memiliki korelasi paling tinggi dengan **jumlah_armroll** (0.61), menunjukkan bahwa peningkatan jumlah sampah cenderung berkaitan erat dengan peningkatan jumlah arm roll yang digunakan untuk pengangkutan. Ini menunjukkan arm roll mungkin merupakan alat yang utama dalam pengangkutan sampah pada tahun 2021.
  
2. `jumlah_motor_sampah` memiliki korelasi yang sedang dengan `jumlah_dumptruck` (0.55) dan `jumlah_sampah` (0.44), menunjukkan bahwa meskipun motor sampah digunakan untuk pengangkutan, penggunaannya tidak setinggi arm roll atau dump truck.

3. `jumlah_dumptruck` memiliki korelasi kuat dengan `jumlah_sampah` (0.55) dan `jumlah_armroll` (0.58), yang berarti dump truck juga sering digunakan bersamaan dengan arm roll dalam pengangkutan sampah. Dump truck tampaknya memiliki peran signifikan dalam pengangkutan sampah, bersama dengan arm roll.

4. `jumlah_armroll` tidak hanya memiliki korelasi tinggi dengan `jumlah_sampah` (0.61) tetapi juga dengan `jumlah_dumptruck` (0.58), yang menunjukkan bahwa keduanya mungkin digunakan secara bersamaan untuk pengangkutan sampah di tahun 2021.

Korelasi negatif dengan `jumlah_dumptruck` (-0.11) dan `jumlah_armroll` (-0.033) menunjukkan bahwa alat ini jarang digunakan bersamaan dengan dump truck atau arm roll. Ini bisa berarti bahwa truck sampah tidak seumum alat lain dalam pengangkutan sampah pada tahun tersebut.

6. `jumlah_mobil_pickup` memiliki korelasi sedang dengan `jumlah_dumptruck` (0.46) dan `jumlah_armroll` (0.43), tetapi korelasi dengan variabel lainnya relatif rendah.
